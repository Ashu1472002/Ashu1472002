```python
import pandas as pd

# Replace 'path_to_dataset.csv' with the actual path to your COVID-19 dataset file
dataset_path = '/content/covid_19_clean_complete.csv'

# Read the dataset
df = pd.read_csv(dataset_path)

# Display first 5 rows
print("First 5 rows of the dataset:")
print(df.head())
```

```python
# Check for missing values
missing_values = df.isnull()
print("\nMissing values:")
print(missing_values)
```

```python
# Print dataset information
print("\nDataset information:")
print(df.info())
```

```python
df['Active'] = df['Confirmed'] - df['Deaths'] - df['Recovered']
result = df.groupby('Country/Region')[['Confirmed', 'Deaths', 'Recovered', 'Active']].sum().reset_index()
print(result)
```

```python
# @title Default title text
data = df.groupby(['Country/Region', 'Province/State'])[['Confirmed', 'Deaths',
'Recovered']].max()
pd.set_option('display.max_rows', None)
print(data)
```

```python
data = df.groupby('Country/Region')[['Confirmed', 'Deaths', 'Recovered']].sum().reset_index()
result = data[data['Recovered']==0][['Country/Region', 'Confirmed', 'Deaths', 'Recovered']]
print(result)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('/content/covid_19_clean_complete.csv', usecols = [ 'Country/Region', 'Confirmed', 'Deaths', 'Recovered'])
data['Active'] = data['Confirmed'] - data['Deaths'] - data['Recovered']

r_data = data.groupby(["Country/Region"])[["Deaths", "Confirmed", "Recovered", "Active"]].sum().reset_index()
r_data = r_data.sort_values(by='Deaths', ascending=False)
r_data = r_data[r_data['Deaths']>200]
plt.figure(figsize=(20, 5))
plt.plot(r_data['Country/Region'], r_data['Deaths'],color='red')
plt.plot(r_data['Country/Region'], r_data['Confirmed'],color='green')
plt.plot(r_data['Country/Region'], r_data['Recovered'], color='blue')
plt.plot(r_data['Country/Region'], r_data['Active'], color='black')
plt.title('Total Deaths(>200), Confirmed, Recovered and Active Cases by Country')
plt.show()
```
